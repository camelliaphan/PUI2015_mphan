```python
import pylab as pl
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import geopandas as gp
import statsmodels.api as sm 
import statsmodels.formula.api as smf
import seaborn as sns
sns.set(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1, rc=None)
import csv 
import sys 
import os
%pylab inline
```

```python
citibike = pd.read_csv('201506-citibike-tripdata.csv')
```

```python
citibike.head()
```

```python
male = citibike[citibike['gender'] == 1]

female = citibike[citibike['gender'] == 2]
```

```python
pl.figure(figsize = (12,12))
pl.plot(male['birth year'], male['tripduration'], 'o', label = 'Male Riders')
pl.plot(female['birth year'], female['tripduration'], 'o', label = 'Female Riders')
pl.xlabel('Birth Year', fontsize = 16)
pl.ylabel('Trip Duration (seconds)', fontsize = 16) 
pl.title ('A graph presenting age, gender and trip duration of Citibike users', fontsize = 20)
pl.legend(loc = 'best', fontsize = 14)
pl.savefig('graph.png')
```
